```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"/1950, A portrait of Jawaharlal Nehru, the Prime Minister of India, a___.jpg")

if img is None:
    print("Error: Image could not be loaded. Please ensure the path is correct and accessible in Colab.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis("off")
```

```python
bright = cv2.add(img, np.ones(img.shape, dtype='uint8') * 50)
bright_rgb = cv2.cvtColor(bright, cv2.COLOR_BGR2RGB)

plt.imshow(bright_rgb)
plt.title("Image Addition (Brightness Increased)")
plt.axis("off")
```

```python
dark = cv2.subtract(img, np.ones(img.shape, dtype='uint8') * 100)
dark_rgb = cv2.cvtColor(dark, cv2.COLOR_BGR2RGB)

plt.imshow(dark_rgb)
plt.title("Image Subtraction (Brightness Decreased)")
plt.axis("off")
```

```python
mul = cv2.multiply(img, np.array([1.2]))
mul_rgb = cv2.cvtColor(mul, cv2.COLOR_BGR2RGB)

plt.imshow(mul_rgb)
plt.title("Image Multiplication")
plt.axis("off")
```

```python
div = cv2.divide(img, np.array([1.5]))
div_rgb = cv2.cvtColor(div, cv2.COLOR_BGR2RGB)

plt.imshow(div_rgb)
plt.title("Image Division")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
_, mask = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
plt.imshow(mask, cmap='gray')
plt.title("Binary Mask")
plt.axis("off")
```

```python
bit_and = cv2.bitwise_and(img, img, mask=mask)
bit_and_rgb = cv2.cvtColor(bit_and, cv2.COLOR_BGR2RGB)

plt.imshow(bit_and_rgb)
plt.title("Bitwise AND")
plt.axis("off")
```

```python
bit_or = cv2.bitwise_or(img, img, mask=mask)
bit_or_rgb = cv2.cvtColor(bit_or, cv2.COLOR_BGR2RGB)

plt.imshow(bit_or_rgb)
plt.title("Bitwise OR")
plt.axis("off")
```

```python
bit_xor = cv2.bitwise_xor(img, img, mask=mask)
bit_xor_rgb = cv2.cvtColor(bit_xor, cv2.COLOR_BGR2RGB)

plt.imshow(bit_xor_rgb)
plt.title("Bitwise XOR")
plt.axis("off")
```

```python
bit_not = cv2.bitwise_not(img)
bit_not_rgb = cv2.cvtColor(bit_not, cv2.COLOR_BGR2RGB)

plt.imshow(bit_not_rgb)
plt.title("Bitwise NOT (Negative Image)")
plt.axis("off")
```

```python
resized = cv2.resize(img, (300, 300), interpolation=cv2.INTER_LINEAR)
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)

plt.imshow(resized_rgb)
plt.title("Resized Image")
plt.axis("off")
```

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, matrix, (w, h))
rotated_rgb = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)

plt.imshow(rotated_rgb)
plt.title("Rotated Image (45 Degrees)")
plt.axis("off")
```

```python
matrix_t = np.float32([[1, 0, 50],
                       [0, 1, 30]])

translated = cv2.warpAffine(img, matrix_t, (w, h))
translated_rgb = cv2.cvtColor(translated, cv2.COLOR_BGR2RGB)

plt.imshow(translated_rgb)
plt.title("Translated Image")
plt.axis("off")
```

```python
edges = cv2.Canny(gray, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis("off")
```

```python
blur = cv2.GaussianBlur(img, (7,7), 0)
blur_rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

plt.imshow(blur_rgb)
plt.title("Gaussian Blurred Image")
plt.axis("off")
```

```python
img1 = np.zeros((256, 256), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        if j > i:
            img1[i, j] = 255   # White
        else:
            img1[i, j] = 0     # Black

plt.imshow(img1, cmap='gray')
plt.title("Diagonal Half BlackWhite Image")
plt.axis("off")
```

```python
img2 = np.zeros((256, 256), dtype=np.uint8)

img2[0:128, 0:128] = 0
img2[0:128, 128:256] = 255
img2[128:256, 0:128] = 255
img2[128:256, 128:256] = 0
plt.imshow(img2, cmap='gray')
plt.title("2x2 Quadrant Image")
plt.axis("off")
```

```python
img3 = np.zeros((256, 256), dtype=np.uint8)
block = 32

for i in range(0, 256, block):
    for j in range(0, 256, block):
        if (i//block + j//block) % 2 == 0:
            img3[i:i+block, j:j+block] = 255

plt.imshow(img3, cmap='gray')
plt.title("Chessboard Pattern")
plt.axis("off")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(
    r"/pexels-david-vives-1474251-28990269.jpg",
    0
)
```

```python
r_vals = np.arange(256)
s_vals = np.zeros(256)

r1, r2 = 70, 140
s1, s2 = 30, 200

for i in range(256):
    if r_vals[i] < r1:
        s_vals[i] = (s1 / r1) * r_vals[i]
    elif r_vals[i] <= r2:
        s_vals[i] = ((s2 - s1) / (r2 - r1)) * (r_vals[i] - r1) + s1
    else:
        s_vals[i] = ((255 - s2) / (255 - r2)) * (r_vals[i] - r2) + s2

# Apply the transformation to the image
piecewise = cv2.LUT(img, s_vals.astype(np.uint8))

plt.figure(figsize=(15, 4))


plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")


plt.subplot(1, 3, 2)
plt.imshow(piecewise, cmap='gray')
plt.title("Piecewise Linear Output")
plt.axis("off")


plt.subplot(1, 3, 3)
plt.plot(r_vals, s_vals)
plt.xlabel("Input gray level (r)")
plt.ylabel("Output gray level (s)")
plt.title("rs Transformation")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
r_vals = np.arange(256)
s_vals = np.zeros(256)

r1, r2 = 100, 150

for i in range(256):
    if r1 <= r_vals[i] <= r2:
        s_vals[i] = 255
    else:
        s_vals[i] = 0

# Apply the transformation to the image
slice_no_bg = cv2.LUT(img, s_vals.astype(np.uint8))

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(slice_no_bg, cmap='gray')
plt.title("Gray Level Slicing (No Background)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.plot(r_vals, s_vals)
plt.xlabel("r")
plt.ylabel("s")
plt.title("rs Graph")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
r_vals = np.arange(256)
s_vals = np.zeros(256)

for i in range(256):
    if r1 <= r_vals[i] <= r2:
        s_vals[i] = 255
    else:
        s_vals[i] = r_vals[i]

# Apply the transformation to the image
slice_with_bg = cv2.LUT(img, s_vals.astype(np.uint8))

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(slice_with_bg, cmap='gray')
plt.title("Gray Level Slicing (With Background)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.plot(r_vals, s_vals)
plt.xlabel("r")
plt.ylabel("s")
plt.title("rs Graph")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
T = 120
r_vals = np.arange(256)
s_vals = np.where(r_vals >= T, 255, 0)

# Apply the transformation to the image
threshold_img = cv2.LUT(img, s_vals.astype(np.uint8))

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(threshold_img, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.plot(r_vals, s_vals)
plt.xlabel("r")
plt.ylabel("s")
plt.title("rs Graph")
plt.grid()

plt.tight_layout()
plt.show()
```
